```python
import pandas as pd
import numpy as np
import ast
import os
from clean import clean_and_lemmatize_string
from sklearn.metrics.pairwise import cosine_similarity
pd.set_option('display.max_colwidth', None)
```

```python
from tensorflow.python.client import device_lib
import tensorflow as tf
tf.debugging.set_log_device_placement(True)
gpus = tf.config.list_physical_devices('GPU')
tf.config.set_visible_devices(gpus[0], 'GPU')
```

```python
VOC_PATH = os.path.join("..", "result", "word_freq.csv")
```

# Load the datasets 

```python
dataframe = pd.read_csv(os.path.join("content", "clean_tweet_topics.csv"))[["text", "topics"]]
dataframe2 = pd.read_csv(os.path.join("..", "result", "Trump_tweet_opinion_analytics_with_volting.csv"))[["text", "Opinion"]]
```

```python
dataframe['topics'] = dataframe['topics'].apply(lambda x: ast.literal_eval(x))
vocabulary = pd.read_csv(VOC_PATH, encoding="latin-1")["vocabulary"]
```

# Query

```python
Q = " Donald Trump's presidential campaign"
```

```python
Q_res =clean_and_lemmatize_string(Q)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

vocab_size = len(vocabulary.tolist())
oov_tok = ''
embedding_dim = 300
max_length = 300
padding_type='post'
trunc_type='post'
tokenizer = Tokenizer(num_words = vocab_size, oov_token=oov_tok)
tokenizer.fit_on_texts(dataframe["text"].tolist())

text_sequences = tokenizer.texts_to_sequences([Q_res])
text_padded = pad_sequences(text_sequences, padding='post', maxlen=max_length)
```

# Ranking

```python
model = tf.keras.models.load_model(os.path.join("model","LSTM", "Model", "bidirectional_lstm_NN_ranking.h5"))
```

```python
predictions = model.predict(text_padded)
```

```python
dataframe_res = dataframe2[['text', 'Opinion']]
dataframe_res['score'] = dataframe['topics'].apply(lambda x: cosine_similarity(predictions.reshape((1, -1)), np.array(x).reshape((1, -1))).squeeze())
dataframe_res = dataframe_res.sort_values("score", ascending=False)
```

```python
Q_ =clean_and_lemmatize_string(Q)
```

```python
Q_in = [w for w in Q_.split() if w in vocabulary.tolist()]
indexes = [np.where(pd.Index(vocabulary) == w)[0][0] for w in Q_in]
max_sim_max= []
sim = np.load(SIM_PATH)
sim_ = np.nan_to_num(sim[indexes, :])
for i, index in enumerate(indexes):
    sim_[i, index] = 0
del sim
index = sim_.argsort(axis=1)[:, -K:]
sim2 = np.zeros_like(sim_)
for i in range(index.shape[0]):
    sim2[i, index[i, :]] = sim_[i, index[i, :]]
Q_in.extend(vocabulary[(sim2.sum(axis=0)/K).argsort()[-K:]].tolist())
Q_res = " ".join(Q_in)
print(Q_res)
terms = Q_in
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
vocab_size = len(vocabulary.tolist())
oov_tok = ''
embedding_dim = 300
max_length = 300
padding_type='post'
trunc_type='post'
tokenizer = Tokenizer(num_words = vocab_size, oov_token=oov_tok)
tokenizer.fit_on_texts(dataframe["text"].tolist())

text_sequences = tokenizer.texts_to_sequences([Q_res])
text_padded = pad_sequences(text_sequences, padding='post', maxlen=max_length)
```

```python
predictions = model.predict(text_padded)
```

```python
import cupy as cp
```

```python
mempool = cp.get_default_memory_pool()
pinned_mempool = cp.get_default_pinned_memory_pool()
mempool.free_all_blocks()
pinned_mempool.free_all_blocks()
```

```python
def SIM_GPU(T1, T2):
    sim = cp.dot(T1, T2.T)
    return sim

def SIM2_GPU(T1, T2):
    return cp.sqrt(cp.power(T1,2).sum(axis=1).reshape(T1.shape[0], 1))*cp.sqrt(cp.power(T2,2).sum(axis=1).reshape(T2.shape[0], 1).T)

def SIM(topic_matric, prediction, d=2, dtype=np.float16):
    sim = np.zeros((topic_matric.shape[0], 1), dtype=dtype)
    m = topic_matric.shape[0]//d
    mm = topic_matric.shape[0]%d
    print("Allocation success")
    with cp.cuda.Device(0):
        T1 = cp.asarray(topic_matric, dtype=cp.float16)
        T2 = cp.asarray(prediction, dtype=cp.float16)
        sim = cp.asnumpy(SIM_GPU(T1, T2)/SIM2_GPU(T1, T2))
        mempool.free_all_blocks()
        pinned_mempool.free_all_blocks()
    return sim
```
